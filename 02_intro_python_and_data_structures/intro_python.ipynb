```python
import numpy.ma as ma

data = np.array([1, 2, -999, 4, -999, 6])

# Mask -999 values (treat as invalid)
masked = ma.masked_where(data == -999, data)
print(masked)
# [1, 2, --, 4, --, 6]  (-- represents masked value)

# Operations ignore masked values
print(masked.mean())  # 3.25 (sum 1+2+4+6 = 13 / 4 = 3.25)
print(np.mean(data))  # -163.5 (includes -999!)

# Multiple conditions
data2 = np.array([10, -999, 20, -1, 30, 999])
masked2 = ma.masked_where((data2 == -999) | (data2 < 0) | (data2 > 100), data2)
print(masked2)  # [10, --, 20, --, 30, --]

# Create masked array directly
valid_data = np.array([1, 2, 3, 4, 5])
mask = [False, True, False, True, False]  # True = masked
masked_array = ma.masked_array(valid_data, mask=mask)
print(masked_array)  # [1, --, 3, --, 5]

# Useful methods
print(masked_array.compressed())    # [1, 3, 5] - removes mask
print(masked_array.filled(-1))      # [1, -1, 3, -1, 5] - fills with value
print(masked_array.count())         # 3 (only unmasked)
```

```python
# Create the DataFrame
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'age': [25, 30, 35, 28, 32, 27],
    'salary': [100, 200, 300, 100, 200, 300],
    'city': ["SJC", "SP", "NYC", "SJC", "SP", "NYC"]
})
```

```python
print("Using iterrows():")
for idx, row in df.iterrows():
    print(f"Index: {idx}, Name: {row['name']}, Age: {row['age']}")
print()

# Modifying rows (does NOT modify original - need to use .at or .loc)
for idx, row in df.iterrows():
    df.at[idx, 'age'] = row['age'] + 1  # This works

print(df.head())  # Ages incremented by 1
# NOTE: Avoid iterrows for large DataFrames - it's slow!
```

```python
print("Using itertuples():")
for row in df.itertuples():
    print(f"Index: {row.Index}, Name: {row.name}, Age: {row.age}, Salary: {row.salary}")

# More efficient than iterrows()
for row in df.itertuples(index=False, name='Person'):
    print(f"{row.name} earns ${row.salary}")

# Access by attribute (faster than iterrows)
total = 0
for row in df.itertuples():
    total += row.salary
print(f"Total salary: {total}")
```

```python
# Single condition
mask1 = df['age'] > 28
print(mask1)

# Apply mask to filter rows
older_than_28 = df[mask1]
print(older_than_28)

# Multiple conditions (AND: &)
mask_and = (df['age'] > 25) & (df['city'] == 'NYC')
filtered_and = df[mask_and]
print(filtered_and)

# Multiple conditions (OR: |)
mask_or = (df['age'] < 28) | (df['salary'] > 200)
filtered_or = df[mask_or]
print(filtered_or)

# NOT condition (~)
not_nyc = df[~(df['city'] == 'NYC')]
print(not_nyc)

# Combining with isin()
cities = ['NYC', 'Miami']
filtered_cities = df[df['city'].isin(cities)]
print(filtered_cities)

# String methods with masks
names_starting_a = df[df['name'].str.startswith('A')]
print(names_starting_a)

# Complex mask example
high_earners_under_30 = df[(df['salary'] > 200) & (df['age'] < 30)]
print(high_earners_under_30)

# Using query() method (alternative to masks)
filtered_query = df.query('age > 28 and city == "NYC"')
print(filtered_query)
```

```python
# Concat (empilhar)
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

# Vertical stack
df_concat = pd.concat([df1, df2], axis=0)
print(df_concat)

# Horizontal stack
df_concat_h = pd.concat([df1, df2], axis=1)

# Merge (like SQL JOIN)
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept_id': [101, 102, 101, 103]
})

departments = pd.DataFrame({
    'dept_id': [101, 102, 103],
    'dept_name': ['Sales', 'IT', 'HR'],
    'floor': [3, 2, 1]
})

# Inner join (default)
merged = pd.merge(employees, departments, on='dept_id')
print(merged)

# Left join
left_join = pd.merge(employees, departments, on='dept_id', how='left')

# Right join
right_join = pd.merge(employees, departments, on='dept_id', how='right')

# Outer join
outer_join = pd.merge(employees, departments, on='dept_id', how='outer')

# Join on different column names
df1 = pd.DataFrame({'key': [1, 2], 'value': ['A', 'B']})
df2 = pd.DataFrame({'key2': [1, 2], 'value2': ['C', 'D']})
merged_diff = pd.merge(df1, df2, left_on='key', right_on='key2')
```

```python
df = pd.DataFrame({
    'name': ['alice', 'bob', 'charlie'],
    'salary': [50000, 60000, 70000],
    'bonus': [0.10, 0.15, 0.20]
})

# apply() - apply function to columns/rows
df['salary_uppercase'] = df['name'].apply(lambda x: x.upper())
print(df)

# Apply to entire column
df['total_comp'] = df.apply(lambda row: row['salary'] * (1 + row['bonus']), axis=1)

# map() - replace values
df['name_clean'] = df['name'].map({'alice': 'Alice', 'bob': 'Bob', 'charlie': 'Charlie'})

# Apply to every element
df_numeric = df[['salary', 'bonus']].applymap(lambda x: x * 1.1)
print(df_numeric)

# Vectorized string methods (faster than apply)
df['name_upper'] = df['name'].str.upper()
df['name_len'] = df['name'].str.len()
```

```python
df = pd.DataFrame({
    'date': ['2024-01', '2024-01', '2024-01', '2024-02', '2024-02', '2024-02'],
    'product': ['A', 'B', 'A', 'A', 'B', 'A'],
    'region': ['North', 'North', 'South', 'North', 'South', 'South'],
    'sales': [100, 150, 200, 120, 180, 220],
    'profit': [20, 30, 40, 25, 35, 45]
})

# Pivot table
pivot = pd.pivot_table(df, 
                       values='sales', 
                       index='product', 
                       columns='region', 
                       aggfunc='sum')
print(pivot)
# region  North  South
# product             
# A          220    420
# B          150    180

# Multiple aggregations
pivot_multi = pd.pivot_table(df,
                             values=['sales', 'profit'],
                             index='product',
                             columns='region',
                             aggfunc=['sum', 'mean'])

# With margins (totals)
pivot_margins = pd.pivot_table(df,
                               values='sales',
                               index='product',
                               columns='region',
                               aggfunc='sum',
                               margins=True,
                               margins_name='Total')

# Crosstab (for frequency)
df_cat = pd.DataFrame({
    'gender': ['M', 'F', 'M', 'F', 'M', 'F'],
    'product': ['A', 'A', 'B', 'B', 'A', 'B']
})
crosstab = pd.crosstab(df_cat['gender'], df_cat['product'])
print(crosstab)
# product  A  B
# gender       
# F        1  2
# M        2  1
```

```python
# Convert to datetime
df = pd.DataFrame({
    'date_str': ['2024-01-15', '2024-02-20', '2024-03-25'],
    'value': [10, 20, 30]
})

df['date'] = pd.to_datetime(df['date_str'])
print(df['date'].dt.year)     # 2024, 2024, 2024
print(df['date'].dt.month)    # 1, 2, 3
print(df['date'].dt.day)      # 15, 20, 25
print(df['date'].dt.dayofweek) # 0=Monday, 1=Tuesday...

# Date ranges
date_range = pd.date_range(start='2024-01-01', end='2024-01-10', freq='D')
print(date_range)

# Resampling time series
ts = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=10, freq='H'),
    'value': range(10)
})
ts.set_index('date', inplace=True)
hourly_mean = ts.resample('2H').mean()
daily_sum = ts.resample('D').sum()

# Shift and lag
df['value_shifted'] = df['value'].shift(1)  # Previous row
df['value_diff'] = df['value'].diff()        # Difference
```

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'salary': [55000, 75000, 60000, 52000, 80000],
    'age': [25, 30, 35, 28, 32]
})

# Sort by one column
df_sorted = df.sort_values('salary', ascending=False)
print(df_sorted)

# Sort by multiple columns
df_sorted_multi = df.sort_values(['age', 'salary'], ascending=[True, False])

# Rank (assign rank position)
df['salary_rank'] = df['salary'].rank(ascending=False)
df['dense_rank'] = df['salary'].rank(method='dense', ascending=False)
print(df)

# Reset index after sorting
df_sorted_reset = df.sort_values('salary').reset_index(drop=True)
```

```python
df = pd.DataFrame({
    'category': ['A', 'B', 'A', 'C'] * 10000,
    'value': range(40000)
})

# Check memory usage
print(df.memory_usage(deep=True))
print(df.info(memory_usage='deep'))

# Convert to categorical (saves memory)
df['category'] = df['category'].astype('category')
print(df['category'].cat.categories)  # ['A', 'B', 'C']

# Downcast numeric types
df['value'] = pd.to_numeric(df['value'], downcast='integer')

# Query vs boolean indexing (query is often faster for large data)
result1 = df[df['category'] == 'A']  # Boolean indexing
result2 = df.query('category == "A"')  # Query (more readable for complex)
```
