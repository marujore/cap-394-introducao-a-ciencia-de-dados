```python
url = "https://data.inpe.br/bdc/wtss/v4/wtss/S2-16D-2/timeseries"

payload = {
    "applyAttributeScale": True,
    "attributes": ["NDVI"],
    "start_datetime": "2025-01-01T00:00:00Z",
    "end_datetime": "2025-12-31T23:59:59Z",
    "geom": {
        "type": "Point",
        "coordinates": [
            -49.43414416473638,
            -11.819838188733794
        ]
    },
    "pixelCollisionType": "center"
}

r = requests.post(url, json=payload)

print(r.status_code)

r.json()
```
